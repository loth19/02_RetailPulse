```python
import sqlite3
```

```python
conn = sqlite3.connect("../src/data/02_RetailPulse.db")
```

```python
df_CA_mensuel=pd.read_sql("select round(sum(montant_total),2) as Chiffre_affaire , strftime('%Y-%m', date_commande) as Mois from commandes GROUP BY strftime('%Y-%m', date_commande) ORDER BY strftime('%Y-%m', date_commande) ASC;",conn)
df_CA_mensuel
```

```python
df_top_produits=pd.read_sql("select produits.produit_id,produits.nom,produits.categorie,round(sum(commandes.montant_total),2) as revenue from produits left join commandes on produits.produit_id = commandes.produit_id group by produits.produit_id ORDER BY revenue DESC LIMIT 5;",conn)
df_top_produits
```

```python
clients_vip=("""select clients.client_id,
       clients.nom,
       clients.ville,
       round(COALESCE(sum(commandes.montant_total),0),2) as depenses 
from clients
left join commandes
on clients.client_id = commandes.client_id
group by clients.client_id
ORDER BY depenses DESC
LIMIT 10;""")
df_clients_vip=pd.read_sql_query(clients_vip,conn)
df_clients_vip
```

```python
periodes_creuses=("""select COUNT(*) as nbre_commandes,strftime('%Y-%m', date_commande) as mois 
from commandes
GROUP BY mois 
ORDER BY nbre_commandes ASC
LIMIT 3;
""")
df_periodes_creuses=pd.read_sql_query(periodes_creuses,conn)
df_periodes_creuses
```
