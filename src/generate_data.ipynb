```python
# Etape 1: execution du bloc de code
# Bloc 1: etape de traitement
from datetime import datetime, timedelta
import pandas as pd
import random
from faker import Faker
fake=Faker('fr_FR')
```

```python
# Etape 2: execution du bloc de code
# Bloc 2: etape de traitement
clients=[
    {
        "client_id":i,
        "nom":fake.name().capitalize(),
        "email":fake.email() if random.random()>0.05 else None ,
        "ville":fake.city(),
        'date_inscription':str(fake.date_between(
            start_date=datetime.now()-timedelta(days=365),
            end_date= datetime.now())
        )
    }
    for i in range(1,201)
]

```

```python
# Etape 3: execution du bloc de code
# Bloc 3: etape de traitement
categories=["Telephonie","Informatique",'Audio','Photo','Tv']
produits=[
    {
        'produit_id':i,
        'nom': fake.word() + str(random.randint(6,17)),
        'categorie':random.choice(categories),
        'prix': random.uniform(150.00 , 1999.99)
    }
    for i in range(1,51)
]
```

```python
# Etape 4: execution du bloc de code
# Bloc 4: etape de traitement
commandes=[]
for i in range(1,1001):
    client=random.choice(clients)
    produit=random.choice(produits)
    commandes.append({
        'commande_id':i,
        'client_id':client['client_id'],
        'produit_id':produit['produit_id'],
        "quantite":random.randint(10,100),
        "date_commande":str(fake.date_between(
            start_date=datetime.now() - timedelta(days=365),
            end_date='today')),
        "montant_total":random.uniform(1999.99 , 5000) if random.random()<0.04 else random.choice([-2000.99,0,999999])
    })

doublons= random.sample(commandes,20)
commandes.extend(doublons)
```

```python
# Etape 5: execution du bloc de code
# Bloc 5: etape de traitement
df_clients=pd.DataFrame(clients)
df_produits=pd.DataFrame(produits)
df_commandes=pd.DataFrame(commandes)
```

```python
# Etape 6: execution du bloc de code
# Bloc 6: etape de traitement
df_clients.to_csv('data/raw/clients.csv',index=False)
df_produits.to_csv('data/raw/produits.csv',index=False)
df_commandes.to_csv('data/raw/commandes.csv',index=False)
```
