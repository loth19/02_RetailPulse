# RetailPulse - Nettoyage des donnees
Notebook de nettoyage des tables clients, produits et commandes.

## 1) Chargement des donnees

```python
# Charger les CSV bruts
import pandas as pd
df_clients = pd.read_csv('../clients.csv')
df_produits = pd.read_csv('../produits.csv')
df_commandes = pd.read_csv('../commandes.csv')
df_clients.shape, df_produits.shape, df_commandes.shape
```

## 2) Nettoyage

```python
# Nettoyer les tables principales
df_clients = df_clients.drop_duplicates().copy()
df_clients['email'] = df_clients['email'].fillna('inconnue')
df_commandes = df_commandes.drop_duplicates().copy()
df_commandes = df_commandes[df_commandes['montant_total'] > 0]
df_commandes['date_commande'] = pd.to_datetime(df_commandes['date_commande'], errors='coerce')
```

## 3) Export des donnees propres

```python
# Exporter les CSV nettoyes
import os
os.makedirs('data/clean', exist_ok=True)
df_clients.to_csv('data/clean/clients.csv', index=False)
df_commandes.to_csv('data/clean/commandes.csv', index=False)
df_produits.to_csv('data/clean/produits.csv', index=False)
print('Export termine')
```
